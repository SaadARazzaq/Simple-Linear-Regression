```python
# Changing the names of the columns of dataset to X and Y

cols = ['X', 'Y']
df = pd.read_csv("data_1d.csv", names=cols)
```

```python
# Store X column in 'X' variable and Y column in 'Y' variable
X = df['X'].values
Y = df['Y'].values
```

```python
# Plot it
plt.scatter(X,Y)
plt.show();
```

```python
# Apply equations from above to calculate a and b

# Calculate Denomenator
denominator = X.dot(X) - X.mean() * X.sum()
```

```python
# Calculate numerator and divide by denominator
a = (X.dot(Y) - Y.mean() * X.sum()) / denominator
b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y)) / denominator
```

```python
# Calculate the slope
Y_hat = a*X + b
```

```python
# Plot the whole thing
plt.scatter(X,Y)
plt.plot(X, Y_hat, 'r')
plt.show()
```

### **Determine how good the model is**

---



We calculate the $R^2$ for it and the formula is as follows:
image.png

```python
d1 = Y - Y_hat
d2 = Y - Y.mean()
SS_res = np.dot(d1, d1)
SS_total = np.dot(d2, d2)
R_square = 1 - (SS_res / SS_total)
```

```python
print("The R squared is: ",R_square)
```
